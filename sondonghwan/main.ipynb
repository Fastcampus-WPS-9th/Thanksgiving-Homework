```python
import random

class Player():
    def __init__(self, name, score):
        self.name = name
        self.score = score

    
    
def playGame():
    
    num_list = random.sample(range(1, 10), 3)
    strike = 0
    ball = 0
    count = 0
    name = input("이름을 입력하세요 : ")
    
    print("숫자 야구 게임")
    print("20번 안에 정답을 맞추세요")
    while (strike < 3):
        strike = 0
        ball = 0
        my_num1 = int(input("First num : "))
        my_num2 = int(input("Second num : "))
        my_num3 = int(input("Third num : "))
        
        my_list = [my_num1,my_num2,my_num3]
        
#         print(num_list)
        for i in range(0, 3):
            for j in range(0, 3):
                if(num_list[i] == my_list[j] and i == j):
                    strike += 1
                elif(num_list[i] == my_list[j] and i != j):
                    ball += 1
        print("―――――――――――――――――――――――――――――――――――――――――――――――――")
        print(f'Result :  Strike : {strike} Ball : {ball}')
        print("―――――――――――――――――――――――――――――――――――――――――――――――――")
        count += 1
        if count > 19:
            print("Game Over")
            break
    if count < 20:
        print(f'The number of times you tried : {count}')
        p1 = Player(name, count)
        f = open('ranking.txt', 'at')
        f.write(f'이름 : {p1.name} 시도횟수 : {p1.score}\n')
        f.close
        
def checkRank():
    f = open('ranking.txt', 'rt')
    lines = f.readlines()
    f.close()
    for line in lines:
        print(line)
        
        
def runGame():
    f = open('ranking.txt', 'at')
    f.close
    while True:
        choice = input('= Number Baseball =\n'
	'  1: Play Game\n'
	'  2: Check Ranking\n'
	'  0: Exit\n'
	'    Input : ')
        if choice == '0':
            break
        elif choice == '1':
            playGame()
        elif choice == '2':
            checkRank()
        else:
            print('Choice not exist')
        print("―――――――――――――――――――――――――――――――――――――――――――――――――")

    print('= Turn off game =')

if __name__ == '__main__':
    runGame()
```

```python
import pygame
import random

pygame.init() 
screen = pygame.display.set_mode((500, 600)) 
pygame.display.set_caption("Avoid Rain")

clock = pygame.time.Clock()
run = True
key = None
start_pos = [0, 0]
rain_speed = 5
r_count = 0
class Player():
    pos = [0, 0]
    exists = 0
    def __init__(self, ix, iy, exists):
        self.pos[0] = ix
        self.pos[1] = iy
        self.exists

class Rain():
    pos = [0, 0]
    exists = 0
    def __init__(self, ix, iy, exists):
        self.pos[0] = ix
        self.pos[1] = iy
        self.exists
        
p1 = Player(230, 550, 1)

r1 = Rain(100, 0, 1)
r2 = Rain(0, 0, 1)
r3 = Rain(0, 0, 1)
r4 = Rain(0, 0, 1)
r5 = Rain(0, 0, 1)
r6 = Rain(0, 0, 1)
r7 = Rain(0, 0, 1)

r1.pos[0] = random.randrange(0, 490)
r2.pos[0] = random.randrange(0, 490)
rain_list=[r1, r2, r3, r4, r5, r6, r7]
# 게임 루프
while run:
    # 1) 사용자 입력 처리
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            run = False

    screen.fill(pygame.color.Color(255, 255, 255))
    
    
    
    
    # 플레이어 draw
    pygame.draw.ellipse(screen,
            pygame.color.Color(255, 0, 0), 
            pygame.Rect(p1.pos, (20, 20)))
    
    # 비 draw
    pygame.draw.rect(screen,
            pygame.color.Color(0, 0, 255), 
            pygame.Rect(r1.pos, (3, 10)))
    
    pygame.draw.rect(screen,
            pygame.color.Color(0, 0, 255), 
            pygame.Rect(r2.pos, (3, 10)))
        
    keys = pygame.key.get_pressed() 
    if keys[pygame.K_LEFT] and p1.pos[0] > 0:
        p1.pos[0] -= 4
    if keys[pygame.K_RIGHT] and p1.pos[0] < 480:
        p1.pos[0] += 4

    for rain in rain_list:
        rain.pos[1] += rain_speed
        
    
    pygame.display.flip()    
    clock.tick(10)

pygame.quit()
```
